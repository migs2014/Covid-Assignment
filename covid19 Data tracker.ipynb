```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned CSV file
df = pd.read_csv("owid-covid-data.csv")

# Ensure population and total vaccinations are numeric
df["population"] = pd.to_numeric(df["population"], errors="coerce")
df["total_vaccinations"] = pd.to_numeric(df["total_vaccinations"], errors="coerce")

# Calculate the percentage of vaccinated population
df["vaccination_rate"] = (df["total_vaccinations"] / df["population"]) * 100

# Plot vaccination rate over time for each country
plt.figure(figsize=(10, 6))

for country in df["location"].unique():
    country_data = df[df["location"] == country]
    plt.plot(country_data["date"], country_data["vaccination_rate"], label=country)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Vaccination Rate (%)")
plt.title("Vaccination Rate Over Time for Selected Countries")
plt.legend()
plt.xticks(rotation=45)
plt.grid()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned CSV file
df = pd.read_csv("cleaned_filtered_countries.csv")

# Convert 'date' column to datetime format
df["date"] = pd.to_datetime(df["date"])

# Create a figure with subplots for different metrics
fig, ax = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Plot total cases over time for each country
for country in df["location"].unique():
    country_data = df[df["location"] == country]
    ax[0].plot(country_data["date"], country_data["total_cases"], label=country)

ax[0].set_ylabel("Total Cases")
ax[0].set_title("Total Cases Over Time")
ax[0].legend()
ax[0].grid()

# Plot total deaths over time for each country
for country in df["location"].unique():
    country_data = df[df["location"] == country]
    ax[1].plot(country_data["date"], country_data["total_deaths"], label=country)

ax[1].set_ylabel("Total Deaths")
ax[1].set_title("Total Deaths Over Time")
ax[1].legend()
ax[1].grid()

# Plot total vaccinations over time for each country
if "total_vaccinations" in df.columns:
    for country in df["location"].unique():
        country_data = df[df["location"] == country]
        ax[2].plot(country_data["date"], country_data["total_vaccinations"], label=country)

    ax[2].set_ylabel("Total Vaccinations")
    ax[2].set_title("Total Vaccinations Over Time")
    ax[2].legend()
    ax[2].grid()
else:
    print("Column 'total_vaccinations' not found in the dataset.")

# Customize the x-axis
plt.xlabel("Date")
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the cleaned CSV file
df = pd.read_csv("owid-covid-data.csv")

# Convert 'date' column to datetime format
df["date"] = pd.to_datetime(df["date"])

# Get the latest data for each country
latest_df = df.loc[df.groupby("location")["date"].idxmax()]

# Ensure necessary columns exist
if "iso_code" in latest_df.columns and "total_cases" in latest_df.columns and "total_vaccinations" in latest_df.columns:
    
    # Plot total cases choropleth map
    fig_cases = px.choropleth(
        latest_df,
        locations="iso_code",
        color="total_cases",
        hover_name="location",
        color_continuous_scale="Reds",
        title="Global COVID-19 Case Density"
    )
    fig_cases.show()

    # Plot vaccination rate choropleth map
    latest_df["vaccination_rate"] = (latest_df["total_vaccinations"] / latest_df["population"]) * 100
    fig_vaccinations = px.choropleth(
        latest_df,
        locations="iso_code",
        color="vaccination_rate",
        hover_name="location",
        color_continuous_scale="Blues",
        title="Global COVID-19 Vaccination Rate (%)"
    )
    fig_vaccinations.show()
else:
    print("Error: Required columns ('iso_code', 'total_cases', 'total_vaccinations') not found in dataset.")
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("owid-covid-data.csv")
df["date"] = pd.to_datetime(df["date"], errors="coerce")

latest_df = df.sort_values("date").drop_duplicates("location", keep="last")

fig_cases = px.choropleth(
    latest_df,
    locations="iso_code",
    color="total_cases",
    hover_name="location",
    color_continuous_scale="Reds",
    title="Global COVID-19 Case Density"
)
fig_cases.show()
```
